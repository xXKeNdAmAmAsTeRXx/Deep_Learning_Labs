# Moels Training

## Getting Data and optuna Parameters

```python
import os
import pandas as pd

data_path = os.path.join('Data','trainVal.csv')
data = pd.read_csv(data_path, index_col=False)
X_trainval, y_trainval = data.drop('Air_Quality', axis=1), data['Air_Quality']
y_trainval = pd.get_dummies(y_trainval) # one hot encoding - no bias, we don't assume anything about data

studies = sorted(os.listdir('optuna_results'))

print('Available studies: ')
for s in studies:
    s = s.replace('(', '_')
    s = s.replace(')', '_')
    parsed_s = s.split('_')
    print(f'{parsed_s[1]}/{parsed_s[2]}/{parsed_s[3]} - {parsed_s[4]}')

last_study_path = os.path.join('optuna_results', studies[-1])
last_study = pd.read_csv(last_study_path)


last_study
```

## Best Model

```python
import torch
from utils.training import create_model
from torchview import draw_graph

import matplotlib.pyplot as plt



best_model = create_model(last_study.iloc[0], (X_trainval.shape[1], y_trainval.shape[1]))

model_graph = draw_graph(best_model, input_size=(1, X_trainval.shape[1]))
model_graph.visual_graph.graph_attr.update(rankdir='LR')
model_graph.visual_graph.graph_attr.update(label='Best Model Architecture', labelloc='t', fontsize='20')
model_graph.visual_graph
```

# Training 3 Best Models
Using Custom Writer Dict

```python
from torch.utils.tensorboard import SummaryWriter
from utils.training import create_training_dict
from utils.dir_managment import clean_dir

os.makedirs('Models', exist_ok=True)

log_dir = os.path.join('runs', 'training')
clean_dir(log_dir)
os.makedirs(log_dir, exist_ok=True)

training_dicts = []
for index, row in last_study.iloc[:3].iterrows():
    name = f'Top_{index+1}'
    model_path = os.path.join('Models', name)
    os.makedirs(model_path, exist_ok=True)
    writer = SummaryWriter(os.path.join(log_dir, name))

    t_dict = create_training_dict(row, data=X_trainval, target=y_trainval, write_model_dir=model_path, writer=writer)
    training_dicts.append(t_dict)

training_dicts[0]
```

```python
from utils.training import train_from_dict

for idx, d in enumerate(training_dicts):
    print(f'\n\nTraining {idx} best model\n\n')
    train_from_dict(training_dict=d)
```
