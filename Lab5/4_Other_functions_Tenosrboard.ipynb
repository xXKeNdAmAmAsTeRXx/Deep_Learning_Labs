# Init Writer

```python
# TensorBoard Panel for this task
# tensorboard --logdir Lab5/runs/4_funcs
```

```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter('runs/4_funcs')
```

## Add Image

```python
import torch

img = torch.rand([3, 512,512])
writer.add_image('img',img,0)
```

## Add Figure

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,256)
y = 0.5 * x + 1 + np.random.random(256)

fig = plt.figure(figsize=(8, 6))
plt.scatter(x,y)

writer.add_figure('figure', fig, global_step=0)
```

## Add Histogram

```python
import numpy as np

data = np.random.normal(5,10,256)

writer.add_histogram('histogram', data, global_step=0)
```

```python
writer.close()
```
